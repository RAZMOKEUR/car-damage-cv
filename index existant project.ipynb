```python
# import libraries 

import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import os 

```

```python
# defines directories 

# train img
train_image_paths = "coco-car-damage-detection-dataset/train" 
filenames = os.listdir(train_image_paths)
train_image_paths = [os.path.join(train_image_paths, filename) for filename in filenames]

# train_annotation_paths

train_annotation_paths = "coco-car-damage-detection-dataset/annots/train/" 
filenames = os.listdir(train_annotation_paths)
train_annotation_paths = [os.path.join(train_annotation_paths, filename) for filename in filenames]

# val img
val_image_paths = "coco-car-damage-detection-dataset/val" 
filenames = os.listdir(val_image_paths)
val_image_paths = [os.path.join(val_image_paths, filename) for filename in filenames]

# val_annotation_paths

val_annotation_paths = "coco-car-damage-detection-dataset/annots/val" 
filenames = os.listdir(val_annotation_paths)
val_annotation_paths = [os.path.join(val_annotation_paths, filename) for filename in filenames]
```

```python
import tensorflow as tf
import json

# Load the train images and annotations
train_images = []
train_annotations = []


for image_path, annotation_path in zip(train_image_paths, train_annotation_paths):
  image = tf.keras.preprocessing.image.load_img(image_path)
  image = tf.keras.preprocessing.image.img_to_array(image)
  train_images.append(image)

  with open(annotation_path, 'r') as f:
    annotation = json.load(f)
  train_annotations.append(annotation)

# # Preprocess the train images and annotations
# train_images = [preprocess_image(image) for image in train_images]
# train_annotations = [preprocess_annotation(annotation) for annotation in train_annotations]

# Create the train dataset
train_dataset = tf.data.Dataset.from_tensor_slices((train_images, train_annotations))

# Repeat the same steps for the validation and test sets
val_images = []
val_annotations = []


for image_path, annotation_path in zip(val_image_paths, val_annotation_paths):
  image = tf.keras.preprocessing.image.load_img(image_path)
  image = tf.keras.preprocessing.image.img_to_array(image)
  val_images.append(image)

  with open(annotation_path, 'r') as f:
    annotation = json.load(f)
    val_annotations.append(annotation)

# val_images = [preprocess_image(image) for image in val_images]
# val_annotations = [preprocess_annotation(annotation) for annotation in val_annotations]

val_dataset = tf.data.Dataset.from_tensor_slices((val_images, val_annotations))
```
